```python
def get_sample(data, n_samples=100, random=True, include_clean=True, include_error_transcription=True, repetitions=0.5, prolongations=0.5):
  copy = data.sort_values(by=['speaker', 'chapter', 'file'])

  if not include_clean:
    copy = copy.loc[copy['stutter_count'] > 0]

  if not include_error_transcription:
    copy = copy.loc[copy['transcription_similarity'] > 0.67]

  repetitions = repetitions / (repetitions + prolongations)
  prolongations = 1 - repetitions

  if random:
    return copy.sample(n_samples)
  else:
    return copy.head(n_samples)

def save_sample_to_csv(sample, filename):
  sample.to_csv(BASE_PATH + '/data/' + filename + '.csv', index=False, quoting=csv.QUOTE_NONNUMERIC)
```
